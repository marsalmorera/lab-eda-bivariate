### Part 1: Analyzing Best-Seller Trends Across Product Categories

**Objective**: Understand the relationship between product categories and their best-seller status.

1. **Crosstab Analysis**:
    - Create a crosstab between the product `category` and the `isBestSeller` status.
    
    - Are there categories where being a best-seller is more prevalent? 
    	
    	*Hint: one option is to calculate the proportion of best-sellers for each category and then sort the categories based on this proportion in descending order.*

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('amz_uk_price_prediction_dataset.csv')
```

```python
df.head(3)
```

```python
# crosstab Category. It is the same as counting. We do need to modify as False = 0. And True = 1. 
pd.crosstab(index=df['category'],
            columns='isBestSeller')
```

```python
def categorize_bestseller (i): 
    if i == False:
        return 0 
    else: 
        return 1 
df['bestseller'] = df['isBestSeller'].apply(categorize_bestseller)
df.head(5)
```

```python
crosstab = pd.crosstab(index=df['category'],
                       columns=df['bestseller'] == 1)
crosstab = crosstab[True] #Just to select the ones that are true. 
```

```python
crosstab_results = crosstab.sort_values(ascending=False).head(5).reset_index()
crosstab_results
```

```python
"""
Answer: In this case we are seeing that the categories which are the best sellers and only taking into account the True values are: 

    1.Grocery 
    2.Health & Personal Care
    3.Baby
    4.Sports & Outdoors. 
    5.Power& Hand Tools. 
"""
```

```python
crosstab_result = pd.crosstab(df['category'], df['isBestSeller'])
crosstab_result
```

```python
# Not well displayed, sorting out how to change it. 
crosstab_results.plot(kind='bar')
plt.show()
```

2. **Statistical Tests**:

    - Conduct a Chi-square test to determine if the best-seller distribution is independent of the product category.
    
    - Compute Cramér's V to understand the strength of association between best-seller status and category.

```python
from scipy.stats import chi2_contingency
```

```python
chi2_stats, shi2_pvalue, _, _ = chi2_contingency(crosstab_result)
```

```python
chi2_stats
```

```python
shi2_pvalue
```

3. **Visualizations**:
	- Visualize the relationship between product categories and the best-seller status using a stacked bar chart.

```python
top_5_bestsellers = crosstab_result.sort_values(by=True, ascending=False).head(5)
top_5_bestsellers
```

```python
top_5_bestsellers.plot(kind='bar', stacked=True)
plt.show()
```
